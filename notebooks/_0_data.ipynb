```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Load state boundaries
states = gpd.read_file(
    "https://raw.githubusercontent.com/PublicaMundi/MappingAPI/master/data/geojson/us-states.json"
)

# Derive country border from states (no deprecated API needed)
conus_border = states.dissolve()

fig, ax = plt.subplots(figsize=(12, 7))
states.boundary.plot(ax=ax, linewidth=0.5, color="gray")        # state lines
conus_border.boundary.plot(ax=ax, linewidth=1.2, color="black") # country border

ds.plot_map(color_by="aridity", ax=ax)

ax.set_xlim(-125, -65)
ax.set_ylim(24, 50)
ax.set_title("Basin Aridity across CONUS")
plt.tight_layout()
plt.show()
```
